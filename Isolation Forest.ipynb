```python
#Isolation Forest with box plot

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder

class IsolationForestOutlierDetection:
    def __init__(self, data):
        """
        Initializes the class with the dataset.
        
        Parameters:
            data (pd.DataFrame): The input dataset.
        """
        self.data = data
        self.numeric_data = data.select_dtypes(include='number')  # Only numeric columns are used for outlier detection
        self.outliers = None

    def detect_outliers(self, contamination=0.05):
        """
        Detects outliers using the Isolation Forest algorithm.
        
        Parameters:
            contamination (float): The proportion of the dataset to be considered as outliers.
        
        Returns:
            pd.Series: A boolean series indicating whether each row is an outlier.
        """
        # Fit the Isolation Forest model on numeric data
        model = IsolationForest(contamination=contamination, random_state=42)
        self.data['outlier'] = model.fit_predict(self.numeric_data)

        # -1 indicates outliers, 1 indicates inliers
        self.data['outlier'] = self.data['outlier'].apply(lambda x: 'Outlier' if x == -1 else 'Inlier')
        self.outliers = self.data[self.data['outlier'] == 'Outlier']
        return self.data['outlier']

    def remove_outliers(self):
        """
        Removes outliers from the dataset.
        
        Returns:
            pd.DataFrame: Dataset with outliers removed.
        """
        return self.data[self.data['outlier'] == 'Inlier'].drop(columns=['outlier'])

    def plot_outliers(self):
        """
        Plots a scatter plot of the numeric data with outliers highlighted.
        """
        if self.outliers is None:
            print("Please run detect_outliers() first.")
            return

        plt.figure(figsize=(10, 6))
        for column in self.numeric_data.columns:
            sns.boxplot(data=self.numeric_data[column], color="skyblue")
            plt.title(f'Outliers highlighted for {column}')
            plt.show()


# Example dataset
data = pd.read_csv('Steel_industry_data.csv')

# Instantiate the class
isolation_forest = IsolationForestOutlierDetection(data)

# Detect outliers
isolation_forest.detect_outliers()

# Plot outliers
isolation_forest.plot_outliers()

# Remove outliers
cleaned_data = isolation_forest.remove_outliers()

print("\nCleaned Data (Outliers Removed):")
print(cleaned_data)
```

```python
#Isolation Forest with Scatter plot

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Create a sample dataset
data = pd.read_csv('Steel_industry_data.csv')

class IsolationForestOutlierDetection:
    def __init__(self, data):
        """
        Initializes the class with the dataset.
        
        Parameters:
            data (pd.DataFrame): The input dataset.
        """
        self.data = data

    def detect_outliers(self):
        """
        Detects outliers using the Isolation Forest method.

        Returns:
            pd.DataFrame: Original dataset with an additional column indicating outlier status.
        """
        # Select only numeric columns for outlier detection
        numeric_data = self.data.select_dtypes(include='number')

        # Apply Isolation Forest
        iso_forest = IsolationForest(contamination=0.1, random_state=42)
        outlier_labels = iso_forest.fit_predict(numeric_data)

        # Add outlier status to the original dataset
        self.data['Outlier'] = outlier_labels
        return self.data

    def plot_outliers(self):
        """
        Plots the data points, highlighting outliers in a scatter plot.
        """
        numeric_data = self.data.select_dtypes(include='number')
        plt.figure(figsize=(10, 6))

        # Scatter plot for numeric columns
        for i, column in enumerate(numeric_data.columns):
            plt.scatter(
                self.data.index,
                self.data[column],
                c=np.where(self.data['Outlier'] == -1, 'red', 'white'),
                label=column
            )

        plt.title("Outlier Detection Using Isolation Forest")
        plt.xlabel("Index")
        plt.ylabel("Feature Value")
        plt.legend(loc="upper left")
        plt.show()

    def remove_outliers(self):
        """
        Removes rows identified as outliers from the dataset.

        Returns:
            pd.DataFrame: Dataset with outliers removed.
        """
        cleaned_data = self.data[self.data['Outlier'] != -1].drop(columns=['Outlier'])
        return cleaned_data


# Instantiate the class
outlier_detector = IsolationForestOutlierDetection(data)

# Detect outliers
outlier_data = outlier_detector.detect_outliers()

# Plot the outliers
outlier_detector.plot_outliers()

# Remove outliers
cleaned_data = outlier_detector.remove_outliers()

# Print results
print("Original Data with Outlier Labels:")
print(outlier_data)
print("\nCleaned Data (Outliers Removed):")
print(cleaned_data)
```

```python
#Isolation Forest with tree plot

import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

class MixedDataOutlierDetection:
    def __init__(self, data):
        """
        Initializes the class with the dataset.
        
        Parameters:
            data (pd.DataFrame): The input dataset with mixed data types.
        """
        self.data = data
        self.encoded_data = None
        self.iforest = None

    def preprocess_data(self):
        """
        Encodes categorical variables and scales the dataset for Isolation Forest.
        
        Returns:
            pd.DataFrame: Preprocessed dataset with numeric values.
        """
        encoded_data = self.data.copy()
        label_encoders = {}
        
        # Encode categorical columns
        for column in encoded_data.select_dtypes(include='object').columns:
            le = LabelEncoder()
            encoded_data[column] = le.fit_transform(encoded_data[column])
            label_encoders[column] = le
        
        # Scale the numerical columns
        scaler = StandardScaler()
        numeric_cols = encoded_data.select_dtypes(include=np.number).columns
        encoded_data[numeric_cols] = scaler.fit_transform(encoded_data[numeric_cols])
        
        self.encoded_data = encoded_data
        return encoded_data

    def detect_outliers(self, contamination=0.1):
        """
        Detects outliers using Isolation Forest.
        
        Parameters:
            contamination (float): The proportion of outliers in the dataset.
        
        Returns:
            pd.Series: Series indicating -1 for outliers and 1 for inliers.
        """
        self.iforest = IsolationForest(contamination=contamination, random_state=42)
        predictions = self.iforest.fit_predict(self.encoded_data)
        self.data['Outlier'] = predictions  # Add a new column for outlier detection
        return predictions

    def remove_outliers(self):
        """
        Removes outliers from the dataset based on Isolation Forest results.
        
        Returns:
            pd.DataFrame: Dataset with outliers removed.
        """
        cleaned_data = self.data[self.data['Outlier'] == 1].drop(columns=['Outlier'])
        return cleaned_data

    def plot_isolation_forest(self):
        """
        Visualizes the Isolation Forest tree structure using Decision Tree approximation.
        """
        # Approximate the Isolation Forest with a Decision Tree
        clf = DecisionTreeClassifier(max_depth=3, random_state=42)
        clf.fit(self.encoded_data, self.data['Outlier'])

        # Plot the tree
        plt.figure(figsize=(12, 8))
        plot_tree(clf, filled=True, feature_names=self.encoded_data.columns, class_names=['Outlier', 'Inlier'])
        plt.title("Isolation Forest Tree Approximation")
        plt.show()

# Sample dataset
data = pd.read_csv('Steel_industry_data.csv')

# Instantiate the class
outlier_detection = MixedDataOutlierDetection(data)

# Preprocess the data
preprocessed_data = outlier_detection.preprocess_data()

# Detect outliers
outlier_detection.detect_outliers()

# Plot the Isolation Forest results
outlier_detection.plot_isolation_forest()

# Remove outliers
cleaned_data = outlier_detection.remove_outliers()

# Print cleaned data
print("\nCleaned Data (Outliers Removed):")
print(cleaned_data)
```

```python
class IsolationForestDetector():
    def __init__(self, contamination=0.1, random_state=42):
        self.contamination = contamination
        self.random_state = random_state
        self.model = IsolationForest(contamination=contamination, random_state=random_state)
        self.data = None
        self.predictions = None
        self.column = None

    def load_data(self, file_path, column):
        df = pd.read_csv(file_path)
        print("Dataset loaded successfully!")
        print(df.head())
        #Select specific columns and convert to NumPy array
        self.data = df[column].values.reshape(-1,1)
        self.column =df[column]

    def fit(self):
        self.model.fit(self.data)
        print("Isolation Forest model trained.")

    def predict(self):
        self.predictions = self.model.predict(self.data)
        return self.predictions

    def anomaly_scores(self):
        scores = self.model.decision_function(self.data)
        return scores

    def visualize(self):

        inliers = self.column[self.predictions == 1]
        outliers = self.column[self.predictions == -1]
        
        # Plot boxplots for inliers and original data
        plt.figure(figsize=(10, 6))
        plt.boxplot(self.column, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
        plt.title("Boxplot of Original Data")
        plt.xlabel("Values")
        plt.show()

       

        # Boxplot after removing outliers
        plt.figure(figsize=(10, 6))
        plt.boxplot(inliers, vert=False, patch_artist=True, boxprops=dict(facecolor="lightgreen"))
        plt.title("Boxplot After Removing Outliers")
        plt.xlabel("Values")
        plt.show()

# Example Usage
if __name__ == "__main__":
    # Initialize the detector
    o_detector = IsolationForestDetector(contamination=0.1)

    # Load a dataset (replace 'your_dataset.csv' with your actual dataset file)
    # Example: The dataset should have columns "Feature1" and "Feature2"
    o_detector.load_data("Steel_industry_data.csv", ["Lagging_Current_Power_Factor"])

    # Fit the model
    o_detector.fit()

    # Predict outliers
    predictions = o_detector.predict()
    print("Predictions (1: inlier, -1: outlier):")
    print(predictions)

    # Get anomaly scores
    scores = o_detector.anomaly_scores()
    print("Anomaly Scores:")
    print(scores)

    # Visualize the results
    o_detector.visualize()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data definition at the beginning of the code
data = pd.read_csv('Steel_industry_data.csv')

class OutlierHandler:
    def __init__(self, data):
        """
        Initialize the OutlierHandler with a numeric dataset.
        
        Parameters:
        - data (pd.Series): Input data to process.
        """
        if isinstance(data, pd.Series):
            self.data = data
        else:
            raise TypeError("Data must be a pandas Series.")

    def calculate_iqr(self):
        """Calculate Q1, Q3, and IQR for the dataset."""
        Q1 = self.data.quantile(0.25)
        Q3 = self.data.quantile(0.75)
        IQR = Q3 - Q1
        return Q1, Q3, IQR

    def handle_outliers(self, method='median'):
        """
        Handle outliers in the dataset.
        
        Parameters:
        - method (str): 'median' or 'mean', the method to replace upper-bound outliers.
        
        Returns:
        - pd.Series: Processed dataset with outliers handled.
        """
        Q1, Q3, IQR = self.calculate_iqr()
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        replacement_value = self.data.median() if method == 'median' else self.data.mean()

        # Remove lower-bound outliers
        self.data = self.data[self.data >= lower_bound]

        # Replace upper-bound outliers
        self.data = self.data.apply(lambda x: replacement_value if x > upper_bound else x)

        return self.data

    def plot_boxplot(self, title="Box Plot"):
        """Plot a box plot of the dataset."""
        plt.figure(figsize=(8, 5))
        plt.boxplot(self.data, vert=False)
        plt.title(title)
        plt.xlabel("Value")
        plt.show()

# Create an instance of the OutlierHandler class
handler = OutlierHandler.handle_outliers(data)

print("Original Data:", handler.data.tolist())
handler.plot_boxplot(title="Original Data")

# Handle outliers by replacing upper-bound outliers with the median
processed_data = handler.handle_outliers(method='median')
print("Processed Data:", processed_data.tolist())
handler.plot_boxplot(title="Processed Data")
```
